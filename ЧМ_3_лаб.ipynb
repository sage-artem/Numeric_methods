```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step
```

Формула прямоугольника (левая)

```python
h = []
errors = []
eps = 1e-4
a, b = -1, 1
for n in range(500, 100, -100):
    x = np.linspace(a, b, n)
    step = x[1] - x[0]
    y = x**2 * np.cosh(3*x)
    integral_correct = 22*np.sinh(3) / 27 - 4*np.cosh(3) / 9
    integral_manual = step * np.sum(y)
    error = np.abs(integral_manual - integral_correct)
    h.append(step)
    errors.append(error)
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
plt.title('Формула прямоугольника (левая)')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.plot(h, errors)
np.set_printoptions(precision=3, suppress=True)
errors = np.log(np.array(errors))
h = np.log(np.array(h))
plt.subplot(1, 2, 1)
plt.plot(h, errors)
plt.grid()
plt.title('Формула прямоугольника (левая)')
plt.xlabel('шаг (логарифм)')
plt.ylabel('ошибка (логарифм)')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```

Метод трапеций

```python
h = []
errors = []
eps = 1e-4
a, b = -1.5, 1.5
for n in range(200, 99, -50):
    x = np.linspace(a, b, n)
    step = x[2] - x[1]
    y = np.cosh(np.sin(x))
    integral_correct = 3.759113209485753
    integral_manual = step * (np.sum(y[1:-1]) + (y[0] + y[-1]) / 2)
    error = np.abs(integral_manual - integral_correct)
    h.append(step)
    errors.append(error)
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
plt.title('Метод трапеций')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.plot(h, errors)
np.set_printoptions(precision=3, suppress=True)
errors = np.log(np.array(errors))
h = np.log(np.array(h))
plt.subplot(1, 2, 1)
plt.plot(h, errors)
plt.grid()
plt.title('Метод трапеций')
plt.xlabel('шаг (логарифм)')
plt.ylabel('ошибка (логарифм)')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```

Формула симпсона

```python
h = []
errors = []
eps = 1e-4
a, b = -1, 1
for n in range(400, 99, -50):
    x = np.linspace(a, b, n + 1)
    step = x[1] - x[0]
    y = x**2 * np.cosh(3*x)
    integral_correct = 22*np.sinh(3) / 27 - 4*np.cosh(3) / 9
    ind = np.arange(len(x))
    integral_manual = (step / 3) * (2 * np.sum(y[ind % 2 == 0]) + 4 * np.sum(y[ind % 2 == 1]) - (y[0] + y[-1]))
    error = np.abs(integral_manual - integral_correct)
    h.append(step)
    errors.append(error)
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
plt.title('Формула Симпсона')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.plot(h, errors)
np.set_printoptions(precision=3, suppress=True)
errors = np.log(np.array(errors))
h = np.log(np.array(h))
plt.subplot(1, 2, 1)
plt.plot(h, errors)
plt.grid()
plt.title('Формула Симпсона')
plt.xlabel('шаг (логарифм)')
plt.ylabel('ошибка (логарифм)')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```
