```python
import numpy as np
import matplotlib.pyplot as plt
```

Правая первая производная

```python
a = -5
b = 5
h = []
list_of_errors = []
for n in range(500, 1000, 100):
    x = np.linspace(a, b, n)
    step = x[1] - x[0]
    h.append(step)
    y = np.log(1 + 1 / (1 + x**2))
    y_analit = -2*x / (x**4 + 3 * x**2 + 2)
    y_der_right = (y[1:] - y[:-1]) / step
    y_der_right = np.append(y_der_right, y_analit[-1])
    #error = np.sum((y_der_right - y_analit)**2
    error = max(np.abs(y_der_right - y_analit))
    list_of_errors.append(error)
errors = np.log(np.array(list_of_errors))
h = np.log(np.array(h))
plt.figure()
plt.plot(h, errors)
plt.title('Правая первая производная')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```

Левая первая производная

```python
a = -5
b = 5
h = []
list_of_errors = []
for n in range(500, 1000, 100):
    x = np.linspace(a, b, n)
    step = x[1] - x[0]
    h.append(step)
    y = np.log(1 + 1 / (1 + x**2))
    y_analit = -2*x / (x**4 + 3 * x**2 + 2)
    y_der_left = (y[1:] - y[:-1]) / step
    y_der_left = np.hstack((np.array(y_analit[0]), y_der_left))
    error = max(np.abs(y_der_left - y_analit))
    list_of_errors.append(error)
errors = np.log(np.array(list_of_errors))
h = np.log(np.array(h))
plt.figure()
plt.plot(h, errors)
plt.title('Левая первая производная')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```

Центральная первая производная

```python
a = -5
b = 5
h = []
list_of_errors = []
for n in range(500, 1000, 100):
    x = np.linspace(a, b, n)
    step = x[1] - x[0]
    h.append(step)
    y = np.log(1 + 1 / (1 + x**2))
    y_analit = -2*x / (x**4 + 3 * x**2 + 2)
    y_der_central = (y[2:] - y[:-2]) / (2*step)
    y_der_central = np.append(y_der_central, y_analit[-1])
    y_der_central = np.hstack((np.array(y_analit[0]), y_der_central))
    error = max(np.abs(y_der_central - y_analit))
    list_of_errors.append(error)
errors = np.log(np.array(list_of_errors))
h = np.log(np.array(h))
plt.figure()
plt.plot(h, errors)
plt.title('Центральная первая производная')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.show()
tg = np.mean(np.diff(errors) / np.diff(h))
print(f'Коэффицент наклона прямой = {tg}')
```

Вторая производная (2 порядок)

```python
a = -5
b = 5
errors = []
h = []
for n in range(500, 1000, 100):
    x = np.linspace(a, b, n)
    step = x[1] - x[0]
    h.append(step)
    y = np.log(1 + 1 / (1 + x**2))
    y_analit = (6 * x**4 + 6 * x**2 - 4) / (x**8 + 6 * x**6 + 13 * x**4 + 12 * x**2 + 4)
    y_der2 = (y[2:] - 2*y[1:-1] + y[:-2]) / (step**2)
    y_der2 = np.append(y_der2, y_analit[-1])
    y_der2 = np.hstack((np.array(y_analit[0]), y_der2))
    error = max(np.abs(y_der2 - y_analit))
    errors.append(error)
errors1 = np.log(np.array(errors))
h1 = np.log(np.array(h))
plt.figure()
plt.plot(h1, errors1)
plt.title('Вторая производная (2 порядок)')
plt.xlabel('шаг')
plt.ylabel('ошибка')
plt.show()
tg = np.mean(np.diff(errors1) / np.diff(h1))
print(f'Коэффицент наклона прямой = {tg}')
```
